Обоснование: на основе данных о котировках контракта SPFB.RTS-12.18 (RIZ8)за 2018 год необходимо построить  график "японские свечи" для даты 20 сентября и выполнить прогноз направления графика "японские свечи" на 21 сентября.

```python
import pandas as pd
```

```python
data = pd.read_csv('https://courses.openedu.ru/assets/courseware/v1/824fccdb3e15cab129049a536ece5500/asset-v1:ITMOUniversity+DATANTECH2035+fall_2021_3+type@asset+block/SPFB.RTS-12.18_180901_181231.csv')
data
```

Изменение формата времени: объединение ячеек даты времени и уход от строкового объекта




```python
data['DATE'] = data['<DATE>'] + ' ' + data['<TIME>']
data['DATE'] = pd.to_datetime(data['DATE'], format = "%d/%m/%y %H:%M")
data
```

Выборка по дате 20 сентября 2018.

```python
data_filtered = data[data['DATE'].dt.strftime('%Y-%m-%d') == '2018-09-20']
data_filtered
```

Для удобства чтения дата стала индексом объекта

```python
data_filtered = data_filtered.set_index('DATE')
data_filtered
```

Оставим только необходимые столбцы и переименуем их

```python
data_filtered = data_filtered[['<OPEN>', '<HIGH>', '<LOW>', '<CLOSE>']]
data_filtered.columns = ['open', 'high', 'low', 'close']
data_filtered
```

```python
data_aggregation = data_filtered.resample('1H').agg({'open':'first',
                                                     'high':'max',
                                                     'low':'min',
                                                     'close':'last'})
```

```python
data_aggregation.head()
```

Визуализация агрегированных данных "японские свечи"

```python
import plotly.graph_objects as go
```

```python
figure = go.Figure(data=[go.Candlestick(x=data_aggregation.index,
                                        open=data_aggregation['open'],
                                        high=data_aggregation['high'],
                                        low=data_aggregation['low'],
                                        close=data_aggregation['close'])])
figure.update_layout(
    title='Данные о котировках контракта SPFB.RTS-12.18 (RIZ8) за 20.09.18')
figure.show()
```

Полученные график "японские свечи" за 20 сентября 2018 года в промежутке с 10.00 до 13.00 отображает бычье движение, за которым следует свечной  паттерн разворота "три черных вороны": показатель нисходящего тренда. Опираясь на данные в промежуток с 21.00 до 22.00 можно говорит о наличии разворотного паттерна "нижняя часть пинцета", исходя из которого можно планировать восходящий тренд на 21 сентября 2018г. 

Проверим данное предположение и построим график "японские свечи" за 20 сентября 2018 года.

```python
data_filtered_test_data = data[data['DATE'].dt.strftime('%Y-%m-%d') == '2018-09-21']
data_filtered_test_data = data_filtered_test_data.set_index('DATE')
data_filtered_test_data = data_filtered_test_data[['<OPEN>', '<HIGH>', '<LOW>', '<CLOSE>']]
data_aggregation_test_data = data_filtered_test_data.resample('1H').agg({'<OPEN>':'first',
                                                     '<HIGH>':'max',
                                                     '<LOW>':'min',
                                                     '<CLOSE>':'last'})
import plotly.graph_objects as go
figure_test_data = go.Figure(data=[go.Candlestick(x=data_aggregation_test_data.index,
                                        open=data_aggregation_test_data['<OPEN>'],
                                        high=data_aggregation_test_data['<HIGH>'],
                                        low=data_aggregation_test_data['<LOW>'],
                                        close=data_aggregation_test_data['<CLOSE>'])])
figure_test_data.update_layout(
    title='Данные о котировках контракта SPFB.RTS-12.18 (RIZ8) за 21.09.18')
figure_test_data.show()
```

Опираясь на полученный график можно говорить о том, что выдвинутое предположение верно и в промежутке с 10 до 11.00 наблюдается бычье движенье. В целом, по данным за 21 сентября 2018 года быки отправляют рынок вверх.
